# 02_calidad_y_limpieza – Control de calidad y limpieza de datos

**Objetivo:** Aplicar las reglas de limpieza definidas en el 01_inspeccion_inicial para corregir inconsistencias, manejar valores faltantes, tratar outliers y entregar un dataset limpio listo para el análisis exploratorio (EDA).

---
## Sección 1 – Reglas de limpieza heredadas del 01_inspeccion_inicial

Se listan las reglas definidas en la inspección inicial que se aplicarán en este notebook:

- **Valores faltantes:** Eliminación de filas con >30% de nulos; imputación de valores faltantes en variables clave (mediana para numéricas, moda para categóricas).
- **Outliers:** Detección mediante IQR (rango intercuartílico) y Z-score (>3). Se decide si eliminar, Winsorizar o transformar.
- **Fechas incoherentes:** `last_login_date` < `signup_date` serán corregidas o eliminadas.
- **Valores fuera de rango:** `monthly_watch_time_mins` < 0 o > 1440 (minutos por día) serán tratados como outliers.
- **Estandarización:** Normalización de `country` y `plan_type` a minúsculas y sin espacios extra.

## Sección 2 – Aplicación de reglas de limpieza

### 2.1 Eliminación de duplicados
Se identifican y eliminan registros duplicados basados en `user_id`. Se registra el número de duplicados encontrados y eliminados.

### 2.2 Tratamiento de valores faltantes
- Se calcula el porcentaje de nulos por columna.
- Se eliminan columnas con >30% de nulos.
- Se imputan valores faltantes en variables numéricas con la mediana y en categóricas con la moda.
- Se registra el impacto de cada operación en la retención del dataset.

### 2.3 Tratamiento de outliers
- Se aplica el método IQR para identificar valores atípicos en `monthly_watch_time_mins`.
- Se aplica Winsorización (cap) en los percentiles 1 y 99 para preservar la distribución.
- Se documenta el número de registros afectados y la reducción de la varianza.

### 2.4 Corrección de fechas incoherentes
- Se detectan registros donde `last_login_date` < `signup_date`.
- Se reemplaza `last_login_date` con `signup_date` en estos casos.
- Se documenta el número de correcciones realizadas.

### 2.5 Estandarización de categorías
- Se normaliza `country` a mayúsculas o minúsculas consistentes.
- Se unifican variantes de `plan_type` (ej. 'Premium', 'PREMIUM', 'premium ' → 'premium').
- Se eliminan filas con categorías desconocidas o vacías.
