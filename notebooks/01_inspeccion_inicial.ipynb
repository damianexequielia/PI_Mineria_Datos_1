# 01_inspeccion_inicial – Inspección inicial del dataset

**Objetivo:** Realizar una primera revisión exploratoria del dataset de suscripciones de streaming para identificar su estructura, tipos de variables, valores faltantes, posibles inconsistencias y definir el plan de limpieza y análisis posterior.

---
## Sección 1 – Contexto y descripción de los datos

En esta sección se describe el origen del dataset, el número de filas y columnas, y se listan las variables con sus tipos esperados. Se documenta brevemente el significado de cada variable y su relevancia para el proyecto de minería de datos.

**Variables esperadas:**
- `user_id`: identificador único del usuario.
- `country`: país de residencia del usuario.
- `plan_type`: tipo de plan contratado (ej. free, basic, premium).
- `signup_date`: fecha de registro.
- `last_login_date`: última fecha de actividad.
- `monthly_watch_time_mins`: minutos de visualización mensual.
- `is_active`: indicador de si el usuario está activo.
